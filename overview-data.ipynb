```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Handle data

```python
df = pd.read_csv("./cleveland.csv", header= None)
df.columns = (["age", "sex", "cp", "restbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"])
```

```python
df["target"] = df.target.map({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})
df["thal"] = df.thal.fillna(df.thal.mean())
df["ca"] = df.ca.fillna(df.ca.mean())
```

# Plot value

```python
sns.histplot(data=df, x='age', hue='target', multiple='dodge', shrink=.1)
plt.title('Variation of Age for each target class')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='target', labels=['No Disease', 'Disease'])
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='age', hue='target', data=df , errorbar='sd', capsize=0.05)

plt.title('Distribution of age vs sex with the target class')
plt.xlabel('sex')
plt.ylabel('age')
plt.legend(title='target', labels=['No Disease', 'Disease'])

plt.tight_layout()
plt.show()
```
